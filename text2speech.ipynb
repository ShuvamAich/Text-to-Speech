```python
import numpy as np
import os
import torchaudio
import matplotlib.pyplot as plt
from IPython.display import Audio
from torchaudio.datasets import LIBRISPEECH
```

```python

data_path = "/content/dataset"
if data_path is None:
  train_dataset = LIBRISPEECH(data_path, url="train-clean-100", download=True)
```

```python
waveform, sample_rate, transcript, speaker_id, chapter_id, utterance_id = train_dataset[2]

print("Waveform shape:", waveform.shape)
print("Sample rate:", sample_rate)
print("Transcript:", transcript)
print("Speaker ID:", speaker_id)
print("Chapter ID:", chapter_id)
print("Utterance ID:", utterance_id)
```

```python
plt.figure(figsize=(10, 4))
plt.plot(waveform.t().numpy())
plt.title("Waveform")
plt.show()

Audio(waveform.numpy(), rate=sample_rate)
```
